```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df=df.iloc[:, 5:]
df.head()
```

```python
ap_item1 = apriori(df, min_support=0.15, use_colnames=True)
ap_item1
```

```python
ass_item1 = association_rules(ap_item1,metric="lift", min_threshold=0.7)
ass_item1
```

```python
ass_item1.sort_values('lift',ascending = False)
```

```python
ass_item1.sort_values('lift',ascending = False)[0:10]
```

```python
plt.scatter(ass_item1['support'], ass_item1['confidence'])
plt.show()
```

```python
ap_item2 = apriori(df, min_support=0.25, use_colnames=True)
ap_item2
```

```python
ass_item2 = association_rules(ap_item2,metric="lift", min_threshold=0.9)
ass_item2
```

```python
plt.scatter(ass_item2['support'], ass_item2['confidence'])
plt.show()
```
